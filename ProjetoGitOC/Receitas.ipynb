```python
#importar biblioteca
import pandas as pd
```

```python
#carregar base de dados
df = pd.read_json("C:/Users/arauj/Desktop/OC - Teste/Desafio - DS - OC/receitas.json")
#exibir informações da base de dados
df.head()
```

```python
#ordenar coluna "calories" de forma decrescente
df=df.sort_values(by='calories', ascending=False)
df_calories=df.head(10)
```

```python
#mostrar categorias entre as receitas mais calóricas
df_calories.head(100)
```

```python
#tornar o df em uma matriz (objeto).
calories=df_calories['categories'].to_numpy()
import numpy as np
#selecionar cada palavra exlusiva
calories=np.unique(calories)
print(calories)
```

```python
#transformar a lista de listas em uma única lista
flat_categories = []
for sublist in calories:
    for item in sublist:
        flat_categories.append(item)
#selecionar cada palavra exclusiva
flat_categories=np.unique(flat_categories)
print(flat_categories)
```

```python
#mostrar ingredientes das calorias selecionadas
df_calories['ingredients'].head()
```

```python
#selecionar apenas os strings
df_ingredients= df_calories['ingredients'].astype(str).str.replace('\d+','')
df_ingredients.head()
```

```python
#tornar a lista de ingredientes em um array (matriz ou objeto)
list_ingredients = df_ingredients.to_numpy()
print(list_ingredients)
```

```python
#separar o texto para cada vez que a palavra for encontrada
ingredients =[i.split() for i in list_ingredients]
print(ingredients)
```

```python
#transformar a lista de listas em uma única lista
flat_ingredient = []
for sublist in ingredients:
    for item in sublist:
        flat_ingredient.append(item)
#flat_ingredient=np.unique(flat_ingredient)
print(flat_ingredient)
```

```python
#contar as ocorrências dos ingredientes nas receitas mais calóricas
from collections import Counter
most_common_ingredient=Counter(flat_ingredient)
most_common_ingredient.most_common(50)
```
